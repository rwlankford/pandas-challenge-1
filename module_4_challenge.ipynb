## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

#Create a column for the Total Weight of the Order
df['Total Weight'] = df['unit_weight'] * df['qty']

# Calculate shipping price based on total weight
df['Shipping Price'] = df['Total Weight'] * 7  
df.loc[df['Total Weight'] <= 50, 'Shipping Price'] = df['Total Weight'] * 10 

df
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
sales_tax_rate = 0.0925

calc_sales_tax = (df['subtotal'] + df['Shipping Price']) * sales_tax_rate

df['Total Price'] = df['subtotal'] + df['Shipping Price'] + calc_sales_tax

df
```

```python
# Create a column for the cost of each line using unit cost, qty, and shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['Total Cost'] = df['unit_cost'] * df['qty'] + df['Shipping Price']

df
```

```python
# Create a column for the profit of each line using line cost and line price
df['Profit'] = df['Total Price'] - df['Total Cost']

df
```

```python
# Clean data formats
df['Shipping Price'] = df['Shipping Price'].astype(float)
df['Total Price'] = df['Total Price'].astype(float)
df['Total Cost'] = df['Total Cost'].astype(float)
df['Profit'] = df['Profit'].astype(float)

df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
sorted_qty_df = df.sort_values(by='qty', ascending=False)
highest_qty_client_list = sorted_qty_df['client_id'].head(5).tolist()

client_spend = df[df['client_id'].apply(lambda x: x in highest_qty_client_list)].groupby('client_id')['Total Price'].sum()
client_spend_currency = client_spend.map('${:,.2f}'.format)
client_spend_currency
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. Sort by total profit.

# Filter the DataFrame for the top 5 clients
top_5_client_df = df[df['client_id'].apply(lambda x: x in highest_qty_client_list)]

# Calculate total units purchased, total shipping price, total revenue, and total profit for each client
client_totals = top_5_client_df.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('Shipping Price', 'sum'),
    total_revenue=('Total Price', 'sum'),
    total_profit=('Profit', 'sum')
)

# Sort the summary DataFrame by total profit
summary_df = client_totals.sort_values(by='total_profit', ascending=False)

summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation. Currency should be in millions of dollars.

summary_df = summary_df.rename(columns={
    'total_units_purchased': 'Total Units Purchased',
    'total_shipping_price': 'Total Shipping Price',
    'total_revenue': 'Total Revenue',
    'total_profit': 'Total Profit'
})

summary_df=summary_df.applymap("${:,.2f}M".format)

summary_df
```
